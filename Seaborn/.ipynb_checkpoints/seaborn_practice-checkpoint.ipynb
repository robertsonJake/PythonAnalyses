```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as pp
%matplotlib inline
pp.style.use('seaborn-pastel')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips['tip_pct'] = 100* tips['tip'] / tips['total_bill']
```

```python
tips.head()
```

```python
grid = sns.FacetGrid(tips,row="sex",col='time',margin_titles=True,size=4)
grid.map(pp.hist, 'tip_pct', bins=np.linspace(0,40,15))
```

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot('day','total_bill','sex',data=tips,kind='box',size=8)
    g.set_axis_labels('day','total bill')
```

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=tips, kind='hex')
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg')
```
