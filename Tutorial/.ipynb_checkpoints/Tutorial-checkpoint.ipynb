```python
import pandas as pd                  #imports our pandas toolkit
import numpy as np                   #imports our numpy tookit
from datetime import datetime
import matplotlib.pyplot as pp
%matplotlib inline
```

```python
#we can import data in various types of formats using pandas
#we import it by storing it in a variable.
#pandas will store our file in a DataFrame
births = pd.read_csv('births.csv')
```

```python
#Now lets check out the format of this data, and how it looks in our dataframe.
#We call the head() function with an argument of 10,
#telling it to print the first 10 rows.
births.head(10)
```

```python
#Sometimes your cell names may be ridiculous,
#You can see what they are, and then rename them as such
births.columns
```

```python
#Now let's say I want to look at a specific column or two.
#I can do that by indexing our DataFrame by the column name.
#careful when you do this, the output will be somewhat long.
births['year']
```

```python
#We can look at rows by the row index, remember python is zero index based!
#we have to do this in a dataframe by using the .loc[] method
births.loc[3:5]
```

```python
#or if we want a specific year, we can make a boolean expression in our indexer
births[births['year'] == 1990]
```

```python
#or we can even do other things with this expression such as
#keep in mind that we don't usually have to use & for and, and | for or in python
#but pandas sometimes requires it as our truth operator
births[(births['year'] > 1995) & (births['year'] < 2000)]
```

# So, we can explore this data set. Let's look at decades and compare the amount of each gender born in each decade

```python
#We can create new columns by just doing the same thing as selecting one
#But now we will just set that new column equal to whatever we want!
births['decade'] = 10*(births['year'] // 10)
births.head()
```

```python
#So now we can make a pivot table off of this data
birthsByDecade = births.pivot_table('births', index='decade', columns='gender',
                                    aggfunc='sum')
birthsByDecade.head()
#We can see that it looks a little funny,
#but our decade is our index now instead of 0,1,2,3...
```

```python
#So now let's plot this data!
#We can use either matplotlib or seaborn.
#matplotlib is nice if you want to customize a lot,
#seaborn is nice if you want everything done for you
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
```

```python
birthsByYear = births.pivot_table('births', index='year',
                                  columns='gender', aggfunc='sum')
```

```python
birthsByYear.head()
```

```python
#You can plot pivot tables easily by using the .plot() method on them
birthsByYear.plot()
```

```python
#Or you can do it manually if you wish. Play around
pp.plot(birthsByYear.index,birthsByYear['F'])
pp.plot(birthsByYear.index,birthsByYear['M'])
pp.ylabel('Number of Births / Year')
pp.xlabel('Year')
```
